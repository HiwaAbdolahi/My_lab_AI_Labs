```python
import pandas as pd
import datetime as dt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import warnings





# Leser data fra CSV-filen
df = pd.read_csv('TSLAfull.csv')

# Dropper NaN-oppfringer
df.dropna(inplace=True)

# Konverterer datokolonnen fra streng til flyttall
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].map(dt.datetime.toordinal)

# Dropper undvendige kolonner
df_date_close = df.drop(["Open", "High", "Low", "Adj Close", "Volume"], axis="columns")

# Deler dataene inn i trenings- og testsett
train_data, test_data = train_test_split(df_date_close, test_size=0.2, random_state=42, shuffle=False)

# Skill ut datokolonnen og aksjekurskolonnen for treningsdata
train_date = train_data.drop("Close", axis="columns")
train_close = train_data["Close"]

# Separer datokolonnen og aksjekurskolonnen for testdata
test_date = test_data.drop("Close", axis="columns")
test_close = test_data["Close"]

# Random Forest-regresjon p treningsdata
reg_obj_train = RandomForestRegressor(n_estimators=100, random_state=42)
reg_obj_train.fit(train_date, train_close)
train_prediction = reg_obj_train.predict(train_date)

# Konverterer datodager til dato for treningsdata
train_date_dt = train_date['Date'].apply(dt.datetime.fromordinal)

                                            # Plot for visualisering av treningsdata og Random Forest-regresjon
plt.figure(figsize=(12, 6))
plt.xlabel('Date', fontsize=14)
plt.ylabel('Stockprice', fontsize=14)

# Visualisering av prediksjoner med Random Forest-regresjon
plt.scatter(train_date_dt, train_close, color='blue')
plt.plot(train_date_dt, train_prediction, color='red')

test_prediction = reg_obj_train.predict(test_date)

# Konverterer datodager til dato for testdata
test_date_dt = test_date['Date'].apply(dt.datetime.fromordinal)

                                                    # Plot for visualisering av prediksjoner p testdatasettet
plt.figure(figsize=(12, 6))
plt.xlabel('Date', fontsize=14)
plt.ylabel('Stockprice', fontsize=14)

# Visualisering av prediksjoner med Random Forest-regresjon
plt.scatter(test_date_dt, test_close, color='blue')
plt.plot(test_date_dt, test_prediction, color='red')

# Formaterer inngangsdatoen fr prediksjonsalgoritmen
input_date = dt.date(2019, 1, 1)
input_od = input_date.toordinal()

# Prediksjon av aksjekurs basert p inngangsdatoen
output_train = reg_obj_train.predict([[input_od]])

# Formaterer inngangsdatoen og predikert aksjekurs fr utskrift
formated_date = input_date.strftime("%D")
formated_stockprice = str('%.10f' % output_train[0])

# Utskrift av prediksjonen
outputString = "Predicted stock price for " + formated_date + " is: " + formated_stockprice
print(outputString)

# Feil mellom prediksjonene og faktiske verdier for treningsdata
mse = mean_squared_error(train_close, train_prediction)
r2 = r2_score(train_close, train_prediction)

# Utskrift av feilmlinger
print('Mean squared error for the Training Dataset: %.2f' % mse)
print('Coefficient of determination for the Training Dataset: %.2f' % r2)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(train_close, bins=30, color='blue', alpha=0.5, label='Actual Prices')
plt.hist(train_prediction, bins=30, color='red', alpha=0.5, label='Predicted Prices')
plt.xlabel('Stock Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend()

```

```python
residuals = train_close - train_prediction
plt.figure(figsize=(12, 6))
plt.scatter(train_date_dt, residuals, color='green', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Residuals')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.legend()
```
